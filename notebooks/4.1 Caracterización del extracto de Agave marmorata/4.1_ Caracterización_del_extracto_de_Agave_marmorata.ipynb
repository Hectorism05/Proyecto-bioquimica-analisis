Figura 4.1.2.  Comportamiento del pH en el extracto estabilizado de Agave marmorata / Figura B.1 Validación gráfica del pH mediante script de Python
*texto en cursiva*

```python
import matplotlib.pyplot as plt

# Datos de la Tabla 5 - Parámetro: pH
labels = ['Extracto Agave marmorata']
ph_promedio = [4.03]
error_ph = [0.150] # Desviación estándar

plt.figure(figsize=(6, 8))
# Creamos la barra con el error
plt.bar(labels, ph_promedio, yerr=error_ph, capsize=10, color='#8eb382', edgecolor='#2d5a27', alpha=0.8)

# Configuración estética
plt.ylabel('Potencial de Hidrógeno (pH)', fontsize=12)
plt.ylim(0, 7) # Rango de pH para notar la acidez
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Añadir el valor exacto sobre la barra
for i, v in enumerate(ph_promedio):
    plt.text(i, v + 0.3, f'pH: {v}', ha='center', fontweight='bold')

plt.show()
```

Figura B.2 Validación gráfica de SDT vs conductividad  mediante script de *Python* *texto en cursiva*


```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
parametros = ['SDT (ppm)', 'Conductividad (mS)']
valores = [1410, 1916.97]
errores = [163.97, 83.72]

x_pos = np.arange(len(parametros))

fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(x_pos, valores, yerr=errores, align='center', alpha=0.7, color=['#4da6ff', '#ffcc00'], capsize=8)

ax.set_ylabel('Magnitud de Medición')
ax.set_xticks(x_pos)
ax.set_xticklabels(parametros)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
