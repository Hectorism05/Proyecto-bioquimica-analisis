```python
import matplotlib.pyplot as plt

# Datos de la Tabla 5 - Parámetro: pH
labels = ['Extracto Agave marmorata']
ph_promedio = [4.03]
error_ph = [0.150] # Desviación estándar

plt.figure(figsize=(6, 8))
# Creamos la barra con el error
plt.bar(labels, ph_promedio, yerr=error_ph, capsize=10, color='#8eb382', edgecolor='#2d5a27', alpha=0.8)

# Configuración estética
plt.ylabel('Potencial de Hidrógeno (pH)', fontsize=12)
plt.ylim(0, 7) # Rango de pH para notar la acidez
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Añadir el valor exacto sobre la barra
for i, v in enumerate(ph_promedio):
    plt.text(i, v + 0.3, f'pH: {v}', ha='center', fontweight='bold')

plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# 1. Datos experimentales (Ejemplo con pH por triplicado)
# Asegúrate de que los nombres coincidan exactamente abajo
datos = {
    'F1': [5.4, 5.5, 5.4],
    'F2': [5.6, 5.7, 5.6],
    'F3': [5.5, 5.5, 5.6],
    'F4': [5.8, 5.9, 5.8]
}

# 2. Ejecutar el test de ANOVA
# Corregido: datos['F4'] en lugar de datos['datos_F4']
f_stat, p_value = stats.f_oneway(datos['F1'], datos['F2'], datos['F3'], datos['F4'])

# 3. Interpretación de resultados
print("--- RESULTADOS DEL ANÁLISIS DE VARIANZA (ANOVA) ---")
print(f"Estadístico F: {f_stat:.4f}")
print(f"Valor p (p-value): {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nCONCLUSIÓN: Hay una diferencia significativa (p < 0.05).")
    print("Al menos una de las fórmulas tiene un comportamiento distinto.")
else:
    print("\nCONCLUSIÓN: No hay diferencias significativas (p > 0.05).")
    print("Estadísticamente, todas las fórmulas son iguales en este parámetro.")
```
